# Plot Title And Label

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pluse")
plt.ylabel("Calorie Burnage")

plt.show()
```

# Bars 

```python
x = np.array(["A" ,"B","C","D"])
y = np.array([4,8,12,16])

plt.bar(x,y)

plt.show()
```

# Bar Width 

```python
x = np.array(["A" ,"B","C","D"])
y = np.array([4,8,12,16])

plt.bar(x,y , width = 0.1)

plt.show()
```

```python
x = np.array(["A" ,"B","C","D"])
y = np.array([4,8,12,16])

plt.barh(x,y , color="g")

plt.show()
```

# Pie Chart

# Labels

```python
y = np.array([10,20,30,40,50])
mylabels = ["A","B","C","D","E"]
plt.pie(y ,labels = mylabels)
plt.show()
```

# Legend

```python
y = np.array([10,20,30,40,50])
mylabels = ["A","B","C","D","E"]
plt.pie(y ,labels = mylabels)
plt.legend()
plt.show()
```
