# *args and **kwargs

```python
def total_amount(price1 , price2, price3, price4, price5):
    total = price1 + price2 + price3 + price4 + price5
    return total
```

```python
print('Total order amount:',total_amount(700 ,599,650,900,820))
```

```python
def total_amount(*args):
    total = 0
    for arg in args:
        total += arg
    return total
```

```python
 print('Total order amount:',total_amount(700 ,599,650,900,820))
```

```python
 print('Total order amount:',total_amount(700 ,599,650,900,820,620,450,890,700,500))
```

```python
def total_number(*prices):
    total = 0
    for price in prices:
        total += price
    return total
print('Total order amount :',total_amount(650,450,350,250,400,600,700,800,900,100))
```

```python
def total_discounted_amount(*prices , discount = 0.0):
    total = 0 
    for price in prices:
        total += price
    total_discount_price = total - discount * total
    return total_discount_price
print('Total order amount :',total_discounted_amount(650,450,350,250,400,600,700,800,900,100))
```

```python
def customer_net_spend(*prices , discount = 0.0 , **kwargs):
    total = 0 
    for price in prices:
        total += price
    total_discounted_price = total - discount*total
    net_spend = total_discounted_price - kwargs['cashback']
    return net_spend
additionals = {'cashback' : 5}
print('Customer net spend (during last day of festive season):', customer_net_spend(700, 599, 650, discount=0.05, **additionals))  
```

```python
def order_summary(*prices , **additionals):
    total = 0
    for price in prices:
        total += price
    net_spend = total - additionals['discount'] * total - additionals ['cashback']

    if total > 10000:
        reward_points = 300
    elif total > 5000:
        reward_point = 200
    elif total > 2000:
        reward_point = 100
    else:
        reward_point = 0
    return total , net_spend , reward_point

additionals = {'discount' : 0.5 , 'cashback' : 5}
ta, ns, rp = order_summary(700, 599, 750, **additionals)
print('Customer Order Summary:\n', '\nTotal Amount:', ta, '\nTotal Discounted Amount:', ns, '\nReward Points Earned:', rp)
```

#MAP

```python
numbers = [ 1,2,3,4]
result = list(map(lambda x:x*2, numbers))
print(result)
```

#FILTERS

```python
numbers = [ 1,2,3,4,5,6]
even_numbers = list(filter(lambda x:x%2==0, numbers))
print(even_numbers)
```
