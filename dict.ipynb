# Dictionary 

```python
car_dict={"brand":"BMW","model":"iX","year":1964}
print(car_dict)
```

# Len of dictionary

```python
print(len(car_dict))
```

#Type of data

```python
print(type(car_dict))
```

```python
print(car_dict['brand'])
```

# for duplicate keys

```python
car_dict={"brand":"BMW","model":"iX","year":1964,"year":2000}
print(car_dict)
```

```python
car_dict={"brand":"BMW","model":"iX","year":2024,"year":2000}
print(car_dict)
```

# for accessing dict items

```python
x = car_dict.get("model")
print(x)
```

```python
car_dict.keys()
```

```python
car_dict.values()
```

# to change values of dictinary 

```python
car_dict["year"]=2025
print(car_dict)
```

```python
car_dict["brand"]="maruti"
print(car_dict)
```

```python
car_dict["model"]="suzuki"
print(car_dict)
```

#To add new key:value of dictionary

```python
car_dict["color"]="white"
print(car_dict)
```

#The values in dictionary items can be of any data type 

```python
d3={"name":"sakshi","year":2025,"pass":True,"marks":[70,80,90]}
print(d3)
```

```python
d3.update({"year":2026})
print(d3)
```

```python
d3.update({"email" : "sakshibisht@gmail.com","phoneno":8976789650})
print(d3)
```

#remove items

```python
d3.pop("phoneno")
print(d3)
```

```python
#the popitem() method removes the last inserted items
```

```python
d3.popitem()
print(d3)
```

```python
#del keyword removes the item with the specified key name 
```

```python
del d3["marks"]
print(d3)
```

```python
# clear method 
```

```python
d4={"fruit":"apple","colour":"red"}
print(d4)
```

```python
d4.clear()
print(d4)
```

```python
myfamily = {
    "child1" :{
       "name" : "sakshi",
        "year" : 2004
    },
    "child2" : {
        "name" : "sumit",
        "year" : 2007
    },
    "child3" : {
        "name" : "amit",
        "year" : 2003
    }
}
print(myfamily)
```
